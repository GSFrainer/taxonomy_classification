```python
import pandas as pd
import numpy as np
```

```python
epochs_path = "../results/epochs"
summarized_path = "../results/summarized"
```

```python
tests_ids = ["1729465000","1729558967", "1729606540", "1730077412"]
```

```python
results = []

for id in tests_ids:
    result = pd.read_csv(summarized_path+"/"+id+"_models_train_test.csv",
                         dtype={'id': str},
                         usecols=['id', 'start_time', 'end_time', 'level', 'minimun_entries', 'data_augmentation', 'batch_size', 'epochs', 'model', 'loss_function', 'learning_rate', 'optimizer', 'obs', 'error', 'best_epoch', 'train_acc_best_epoch', 'train_loss_best_epoch', 'test_acc_best_epoch', 'test_loss_best_epoch'])
    result["id"] = (id+"_")+result["id"] 
    results.append(result)
```

```python
results = pd.concat(results).reset_index(drop=True)
results
```

```python
results[
    [
        'id',
        'start_time',
        'end_time',
        'level',
        'minimun_entries',
        'data_augmentation',
        # 'batch_size',
        # 'epochs',
        'model',
        'best_epoch',
        'train_acc_best_epoch',
        'train_loss_best_epoch',
        'test_acc_best_epoch',
        'test_loss_best_epoch'
    ]
].loc[results["end_time"].notna()].sort_values(by=["level", "test_acc_best_epoch", "test_loss_best_epoch"])
```
